# Getting started with Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
!pip install --upgrade seaborn
```

```python
sales = pd.read_excel('sales.xlsx')
sales
```

# Relative plot

```python
sns.relplot('Sales','Profit',data=sales)
plt.show()
```

```python
# Interactivity with Order Priority
sns.relplot('Sales','Profit',data=sales,hue='Order Priority')
plt.show()
```

```python
# Interactivity with Order Priority
sns.relplot('Sales','Profit',data=sales,hue='Order Priority',style='Order Priority') #Style helps to change shapes
plt.show()
```

```python
# Interactivity with Order Priority, style with shipm mode
sns.relplot('Sales','Profit',data=sales,hue='Order Priority',style='Ship Mode') #Style helps to change shapes
plt.show()
```

```python
# Size with Discount
sns.relplot('Sales','Profit',data=sales,hue='Order Priority',size='Discount',sizes=(25,200))
plt.show()
```

# Line chart

```python
sns.relplot('Order Date', 'Sales',data=sales,kind='line')
```

```python
plt.show()
```

```python
# Since there are alot of year values, we need to classify the year values
sales['year'] = sales['Order Date'].dt.year
sales.year.unique()
```

```python
sns.relplot('year','Sales',data=sales,kind='line') 
# Blue line is the trend
# Light blue region is a confidence interval
```

```python
plt.show()
```

```python
sns.relplot('year','Sales',data=sales,kind='line',estimator=None) #to show the actual trend 
```

```python
plt.show()
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode')
```

```python
plt.show()
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',style='Product Category',ci=False)
```

```python
plt.show()
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',col='Product Category',ci=False)
```

```python
plt.show()
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',col='Region',ci=False,col_wrap=3)
```

```python
plt.show()
```

# Facet charts

```python
sns.relplot('Sales','Profit',data=sales,col='Customer Segment')
```

```python
plt.show()
```

# Scatter plot or Swarm plot

```python
sns.catplot("Ship Mode","Sales",data=sales)
```

```python
plt.show()
```

```python
sns.catplot("Ship Mode","Sales",data=sales,jitter=False)
```

```python
plt.show()
```

```python
sns.catplot("Ship Mode","Sales",data=sales.query("Profit > 1000"),jitter=False)
```

```python
plt.show()
```

```python
sns.catplot("Ship Mode","Sales",data=sales.query("Profit > 1000"),kind="swarm")
```

```python
plt.show()
```

```python
sns.catplot("Sales","Ship Mode",data=sales.query("Profit > 1000"),kind="swarm")
```

```python
plt.show()
```

# Violin plot

```python
sns.catplot(x='Product Category',y='Shipping Cost',data=sales,kind='violin')
```

```python
plt.show()
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',hue='Ship Mode',data=sales,kind='violin')
```

```python
plt.show()
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',hue='Ship Mode',inner='stick',data=sales,kind='violin')
```

```python
plt.show()
```

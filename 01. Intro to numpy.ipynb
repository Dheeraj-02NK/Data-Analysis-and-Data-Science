# Find the yeild of apple
yeild_of_apples = w1 * temperature + w2 * rainfall + w3 * humidity


```python
w1,w2,w3 = 0.3, 0.2, 0.5
```

# Data Sample


```python
kanto_temp = 73
kanto_rainfall = 67
kanto_humidity = 43
```

```python
yeild_of_apple = (w1*kanto_temp)+(w2*kanto_rainfall)+(w3*kanto_humidity)
print(f'The expected yeild of apples in kanto region is {yeild_of_apple} tons per hector')
```

```python
kanto = [73,67,43]
johto = [91,88,64]
hoenn = [87,134,58]
sinnoh = [102,43,37]
unova = [69,96,70]
```

```python
weights = [w1,w2,w3]
```

```python
def crop_yeild(region,weights):
    '''This function computer crop yeild of given region
    parameters:
                region as list
                weights as list'''
    result = 0
    for x,w in zip(region,weights):
        result +=x*w
    print(f'The expected yeild of apples in kanto region is {result} tons per hector')
    
```

```python
crop_yeild(kanto,weights)
```

# Comma seperated values(CSV)

```python
import urllib.request
url = 'https://raw.githubusercontent.com/chidanandamurthy/dataAnalysis/main/italy.txt'
urllib.request.urlretrieve(url,'climate.txt')
```

```python
url
```

```python
climate_data = np.genfromtxt('climate.txt',delimiter=',',skip_header=1)
```

```python
print(climate_data)
```

```python
climate_data.shape
```

```python
yeild_of_region = climate_data @ weights
```

```python
print('Yeild of 100th region is ',yeild_of_region[99])
```

```python
climate_result = np.concatenate((climate_data,yeild_of_region.reshape(10000,1)),axis=1)
```

```python
climate_result
```

```python
np.savetxt('climate_result.txt',
          climate_result,
          fmt = '%.2f',
          delimiter=',',
          header='Temperatute,Rainfall,Humidity,Yeild_apples',
          comments='')
```
